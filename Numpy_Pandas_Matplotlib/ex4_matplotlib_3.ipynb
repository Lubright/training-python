```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# ch1_1.py - plot
x = np.arange(9)
y = np.power(x, 2) # https://numpy.org/doc/stable/reference/generated/numpy.power.html
x2 = np.arange(9)
y2 = np.power(x2, 3)

plt.plot(x, y, 
         color='green', 
         marker='o', 
         linestyle="-", 
         linewidth=2,
         label="x^2") # https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot
plt.plot(x2, y2, 
         color='red', 
         marker='^', 
         linestyle="--", 
         linewidth=2, 
         label="x^3") # 第二組數據

plt.xlim(0, 8)
plt.ylim(0, y2.max()+100)

plt.xticks(np.linspace(0, 8, num=17))

plt.title("ch1_1.py", fontsize=24)
plt.xlabel("x", fontsize=14)
plt.ylabel("squre(x)", fontsize=14)
plt.legend(loc="upper left", bbox_to_anchor=(1,1))
plt.text(2, 100, "(2, 100)")

plt.tick_params(axis="both", labelsize=12, color="red")

# plt.tight_layout(pad=7)
plt.tight_layout()
plt.grid(True)


plt.savefig("out1.png")
plt.show()
```
