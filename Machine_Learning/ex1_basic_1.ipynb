## 聯立方程式
### 使用 sympy

```python
from sympy.solvers import solve
from sympy import Symbol
```

```python
# ch6_1.py
a = Symbol("a") # 定義公式中的 a
b = Symbol("b") # 定義公式中的 b

eq1 = a + b - 1
eq2 = 5*a + b - 2

result = solve((eq1, eq2))
print(result, type(result))
```

```python
# ch6_2.py - 使用斜率
## (600, 0), (1000, 12)

x1, y1 = (600, 0)
x2, y2 = (1000, 12)

a = (y2-y1)/(x2-x1)
## y = ax + b
b = 0 - a*600
print(a, b)

x = np.arange(0, 2500, 50)
y = np.multiply(a, x) + b

plt.plot(x1, y1,
         marker='o', 
         linestyle="-", 
         linewidth=2) # https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot

plt.plot(x2, y2,
         marker='o', 
         linestyle="-", 
         linewidth=2) # https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot

plt.plot(x, y, 
         color='green',
         linestyle="-", 
         linewidth=2) # https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot


plt.title("ch6_2.py", fontsize=24)
plt.xlabel("x", fontsize=14)
plt.ylabel("y", fontsize=14)

plt.text(x1-150, y1+3, "f(x1)")
plt.text(x2-150, y2+3, "f(x2)")


plt.tight_layout()
plt.grid(True)

plt.show()
```

```python
# ch6_2.py - 解聯立
## (600, 0), (1000, 12)

x1, y1 = (600, 0)
x2, y2 = (1000, 12)

a = Symbol('a')
b = Symbol('b')

eq1 = a*x1 + b - y1
eq2 = a*x2 + b - y2

result = solve((eq1, eq2))
a = float(result[a])
b = float(result[b])
print(a, b)

x = np.arange(0, 2500, 50)
y = np.multiply(a, x) + b

plt.plot(x1, y1,
         marker='o', 
         linestyle="-", 
         linewidth=2) # https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot

plt.plot(x2, y2,
         marker='o', 
         linestyle="-", 
         linewidth=2) # https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot

plt.plot(x, y, 
         color='green',
         linestyle="-", 
         linewidth=2) # https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot


plt.title("ch6_2_2.py", fontsize=24)
plt.xlabel("x", fontsize=14)
plt.ylabel("y", fontsize=14)

plt.text(x1-150, y1+3, "f(x1)")
plt.text(x2-150, y2+3, "f(x2)")

plt.tight_layout()
plt.grid(True)

plt.show()
```
